```python
import pandas as pd

match_list = []

for type_match in data.get("typeMatches", []):
    for series_matches in type_match.get("seriesMatches", []):
        series_data = series_matches.get("seriesAdWrapper", {})
        series_id = series_data.get("seriesId")
        series_name = series_data.get("seriesName")

        for match in series_data.get("matches", []):
            info = match.get("matchInfo", {})
            match_id = info.get("matchId")
            match_desc = info.get("matchDesc")
            match_format = info.get("matchFormat")
            start_date = info.get("startDate")
            status = info.get("status")
            team1 = info.get("team1", {}).get("teamName")
            team2 = info.get("team2", {}).get("teamName")
            venue = f"{info.get('venueInfo', {}).get('ground')}, {info.get('venueInfo', {}).get('city')}"

            score = match.get("matchScore", {})
            team1_innings = score.get("team1Score", {}).get("inngs1", {})
            team2_innings = score.get("team2Score", {}).get("inngs1", {})

            team1_runs = team1_innings.get("runs", 0)
            team1_wickets = team1_innings.get("wickets", 0)
            team1_overs = team1_innings.get("overs", 0)

            team2_runs = team2_innings.get("runs", 0)
            team2_wickets = team2_innings.get("wickets", 0)
            team2_overs = team2_innings.get("overs", 0)

            match_list.append({
                "series_id": series_id,
                "series_name": series_name,
                "match_id": match_id,
                "match_desc": match_desc,
                "match_format": match_format,
                "start_date": start_date,
                "status": status,
                "team1": team1,
                "team2": team2,
                "venue": venue,
                "team1_runs": team1_runs,
                "team1_wickets": team1_wickets,
                "team1_overs": team1_overs,
                "team2_runs": team2_runs,
                "team2_wickets": team2_wickets,
                "team2_overs": team2_overs
            })

# ✅ Optional: Convert to DataFrame
df = pd.DataFrame(match_list)
print(df)
#match_list

```

```python
import requests
import mysql.connector
import time  # Optional: to avoid hitting rate limits

import re

def clean_capacity(value):
    if not value:
        return 0
    # Extract only digits and commas
    match = re.search(r'\d[\d,]*', value)
    if match:
        return int(match.group(0).replace(',', ''))
    return 0

# Step 1: Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Welcome*1",
    database="cricbuzz"
)
cursor = conn.cursor()

# Step 2: Create table
# cursor.execute("""
# CREATE TABLE IF NOT EXISTS venue_details (
#     venue_id INT PRIMARY KEY,
#     name VARCHAR(100),
#     city VARCHAR(100),
#     country VARCHAR(100),
#     capacity INT,
#     timezone VARCHAR(100)
# )
# """)

# Step 3: Fetch venue list
url = "https://cricbuzz-cricket.p.rapidapi.com/series/v1/11052/venues"
headers = {
    "x-rapidapi-key": "af06ca3954mshac617103945081ap11717ejsn97324ff1b680",
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}
response = requests.get(url, headers=headers)

if response.status_code == 200:
    venues = response.json()

    # Step 4: Loop through venue IDs
    for venue in venues.get('seriesVenue', []):
        venue_id = venue['id']
        detail_url = f"https://cricbuzz-cricket.p.rapidapi.com/venues/v1/{venue_id}"
        detail_response = requests.get(detail_url, headers=headers)

        if detail_response.status_code == 200:
            try:
                detail = detail_response.json()
                cursor.execute("""
                        INSERT INTO venue_details (venue_id, name, city, country, capacity, timezone)
                        VALUES (%s, %s, %s, %s, %s, %s)
                        ON DUPLICATE KEY UPDATE
                        name = VALUES(name),
                        city = VALUES(city),
                        country = VALUES(country),
                        capacity = VALUES(capacity),
                        timezone = VALUES(timezone)
                    """, (
                        venue_id,
                        detail.get('ground', ''),
                        detail.get('city', ''),
                        detail.get('country', ''),
                        clean_capacity(detail.get('capacity', '')),
                        detail.get('timezone', '')
                ))
            except Exception as e:
                print(f"❌ Error parsing venue {venue_id}: {e}")
        else:
            print(f"❌ Failed to fetch venue {venue_id}. Status code: {detail_response.status_code}")

        time.sleep(2)  # Optional: pause to avoid rate limits

    conn.commit()
    print("✅ All venue details inserted successfully.")
else:
    print(f"❌ Failed to fetch venue list. Status code: {response.status_code}")

cursor.close()
conn.close()
```

```python
import requests
import mysql.connector
from datetime import datetime
import time
import re


import re


def extract_winner(status):
    if isinstance(status, str):
        # Case 1: Standard win format → "India won by 5 wickets"
        match = re.match(r"^(.*?)\s+won\b", status, re.IGNORECASE)
        if match:
            return match.group(1).strip()

        # Case 2: Tied match with Super Over → "Match tied (India Won the Super Over)"
        match = re.search(r"\((.*?)\s+Won\b", status)
        if match:
            return match.group(1).strip()

    return None

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Welcome*1",
    database="cricbuzz"
)
cursor = conn.cursor()

# Step 2: Fetch match_ids from recent_matches table
cursor.execute("SELECT match_id FROM recent_matches")
match_ids = [row[0] for row in cursor.fetchall()]

# Step 3: Setup API headers
headers = {
    "x-rapidapi-key": "af06ca3954mshac617103945081ap11717ejsn97324ff1b680",
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

# Step 4: Loop through match_ids and collect scorecard data
all_scorecards = []

for match_id in match_ids:
    url = f"https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/{match_id}/scard"
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        winner_result = response.json()
        scorecard_entries = winner_result.get("scorecard", [])

        for entry in scorecard_entries:
            all_scorecards.append({
                "Match ID": match_id,
                "Innings ID": entry.get("inningsid"),
                "Team": entry.get("batteamname"),
                "Short Name": entry.get("batteamsname"),
                "Score": entry.get("score"),
                "Wickets": entry.get("wickets"),
                "Overs": entry.get("overs"),
                "Run Rate": entry.get("runrate"),
                "Balls": entry.get("ballnbr"),
                "Runs per Ball": entry.get("rpb"),
                "Declared": entry.get("isdeclared"),
                "Follow-on": entry.get("isfollowon"),
                "Match Status": winner_result.get("status")
            })
    else:
        print(f"Failed to fetch scorecard for match_id {match_id}")

# Step 5: Create DataFrame
df = pd.DataFrame(all_scorecards)
# Add Winner column by extracting from Match Status

df["Winner"] = df["Match Status"].apply(extract_winner)

# Step 6: Display the scorecards
#print("\nAll Match Scorecards:")
#print(df)

# Optional: Save to CSV
# df.to_csv("cricbuzz_scorecards.csv", index=False)
# Step 7: Create table if it doesn't exist
cursor.execute("""
    CREATE TABLE IF NOT EXISTS match_scorecards (
        match_id INT,
        innings_id INT,
        team VARCHAR(100),
        short_name VARCHAR(10),
        score INT,
        wickets INT,
        overs FLOAT,
        run_rate FLOAT,
        balls INT,
        runs_per_ball FLOAT,
        declared BOOLEAN,
        follow_on BOOLEAN,
        match_status VARCHAR(255),
        winner VARCHAR(100)
    )
""")

# Step 8: Insert DataFrame rows into the table
for _, row in df.iterrows():
    cursor.execute("""
        INSERT INTO match_scorecards (
            match_id, innings_id, team, short_name, score, wickets, overs,
            run_rate, balls, runs_per_ball, declared, follow_on,
            match_status, winner
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row["Match ID"], row["Innings ID"], row["Team"], row["Short Name"],
        row["Score"], row["Wickets"], row["Overs"], row["Run Rate"],
        row["Balls"], row["Runs per Ball"], row["Declared"], row["Follow-on"],
        row["Match Status"], row["Winner"]
    ))

# Step 9: Commit and close connection
conn.commit()
cursor.close()
conn.close()

print("\n✅ Match scorecards successfully saved to MySQL table 'match_scorecards'.")
```
